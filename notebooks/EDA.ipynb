### Exploratory Data Analysis of Laptop Pricing Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Load the dataset

file = "../data/laptops.csv"  # Relative path from notebooks folder

df = pd.read_csv(file)
print(f"Dataset shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nData types:\n{df.dtypes}")
print(f"\nMissing values:\n{df.isnull().sum()}")
df.head()
```

**Insight**: The dataset contains 238 laptop records with multiple features including categorical (Category, GPU, OS, CPU_core) and continuous variables (Screen_Size_inch, CPU_frequency, Weight_Pounds, RAM_GB, Storage_GB_SSD, Price). This analysis will help identify which features most significantly impact laptop pricing.


#### Step1 - Visualization of Each Feature Patterns


- Continues Features
    1. Screen_Size_inch
    2. CPU_frequency
    3. Weight_pounds

```python
#Regression Plot of these three individual features with the target variable Price to check the correlation between them.
#1. Screen_Size_inch
sns.regplot(x=df["Screen_Size_inch"], y=df["Price"])
plt.ylim(0,)
plt.show()
```

```python

#2. CPU_frequency
sns.regplot(x=df["CPU_frequency"], y=df["Price"])
plt.title("CPU Frequency vs Price Relationship")
plt.xlabel("CPU Frequency")
plt.ylabel("Price ($)")
plt.ylim(0,)
plt.show()
```

```python
# 3. Weight_Pounds
sns.regplot(x=df["Weight_Pounds"], y=df["Price"])
plt.title("Weight vs Price Relationship")
plt.xlabel("Weight (Pounds)")
plt.ylabel("Price ($)")
plt.ylim(0,)
plt.show()
```

```python
#Corelation Matrix
df[["Screen_Size_inch", "Price"]].corr()
for params in ["Screen_Size_inch", "CPU_frequency", "Weight_Pounds"]:
    print(f"Correlation between {params} and Price: {df[[params, 'Price']].corr()}")  
```

**Key Insights from Continuous Feature Analysis:**

1. **CPU Frequency** (Correlation: 0.367): Shows a moderate positive correlation with price. Higher CPU frequencies generally correspond to higher-priced laptops, indicating that processing power is a significant pricing factor.

2. **Screen Size** (Correlation: -0.111): Exhibits a weak negative correlation with price. This suggests that larger screens don't necessarily mean higher prices, possibly due to the prevalence of budget-friendly large-screen laptops or premium compact models.

3. **Weight** (Correlation: -0.050): Shows minimal correlation with price. Weight is not a strong predictor of laptop pricing, indicating that portability (lightweight) doesn't consistently command a premium in this dataset.

### Step3- Pivot Table

```python
# One liner
# df.pivot_table(values="Price", index="GPU", columns="CPU_core", aggfunc=np.mean)

# OR IT can be grouped First the create a Pivot
df_group=df[["GPU", "CPU_core", "Price"]]
df_group=df_group.groupby(["GPU", "CPU_core"], as_index=False).mean()
print("Average Price by GPU and CPU Core Combination:")
print(df_group)
print("\nPivot Table:")
df_pivot=df_group.pivot(index="GPU", columns="CPU_core", values="Price")
df_pivot
```

```python
fig,ax=plt.subplots(figsize=(10, 6))
im=plt.pcolor(df_pivot, cmap="RdBu")
row_labels = df_pivot.columns
col_labels = df_pivot.index

#insert labels
ax.set_xticks(np.arange(df_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(df_pivot.shape[0]) + 0.5, minor=False)
ax.set_xticklabels(row_labels)
ax.set_yticklabels(col_labels)

plt.title("Heatmap: Average Price by GPU and CPU Core Combination")
plt.xlabel("CPU Core")
plt.ylabel("GPU")
plt.colorbar(im, label="Average Price ($)")
plt.tight_layout()
plt.show()
```

**Pivot Table Insights:**

The heatmap reveals the interaction between GPU and CPU core count on pricing:
- **Higher GPU tiers combined with more CPU cores** result in significantly higher average prices
- Certain GPU-CPU combinations show price premiums, indicating these are premium configurations
- The visualization helps identify optimal price-performance combinations and market segments
- Some combinations may be rare or represent specific market niches (gaming, professional workstations)


#### Step 4 - Pearson Correlation and p-values


```python
for param in ['RAM_GB','CPU_frequency','Storage_GB_SSD','Screen_Size_inch','Weight_Pounds','CPU_core','OS','GPU','Category']:
    pearson_coef, p_value = stats.pearsonr(df[param], df['Price'])
    print(param)
    print("The Pearson Correlation Coefficient for ",param," is", pearson_coef, " with a P-value of P =", p_value)
```

**Pearson Correlation & Statistical Significance Insights:**

**Strong Positive Correlations (Statistically Significant):**
1. **RAM_GB (r=0.549, p<0.001)**: Strongest predictor - RAM is the most important pricing factor
2. **CPU_core (r=0.459, p<0.001)**: Number of CPU cores strongly influences price
3. **CPU_frequency (r=0.367, p<0.001)**: CPU speed moderately impacts pricing

**Moderate Positive Correlations:**
4. **GPU (r=0.288, p<0.001)**: GPU type significantly affects price
5. **Category (r=0.286, p<0.001)**: Laptop category is a pricing determinant
6. **Storage_GB_SSD (r=0.243, p<0.001)**: Storage capacity has a moderate positive impact

**Weak/Negative Correlations:**
7. **OS (r=-0.222, p<0.001)**: Operating system shows weak negative correlation (some OS options may be associated with budget models)
8. **Screen_Size_inch (r=-0.111, p=0.089)**: Not statistically significant - screen size doesn't reliably predict price
9. **Weight_Pounds (r=-0.050, p=0.440)**: Not significant - weight is not a pricing factor

**Key Takeaway**: RAM, CPU cores, and CPU frequency are the top three features driving laptop prices, all with strong statistical significance.
