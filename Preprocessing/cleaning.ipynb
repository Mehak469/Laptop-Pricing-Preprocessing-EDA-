```python
#importing Libraries that are used
import pandas as pd
import numpy as np
```

```python
file="laptop_pricing_dataset_mod1.csv"
df=pd.read_csv(file)
df.head()
```

```python
df.info()
df.tail(10)
```

```python
df.drop(columns=["Unnamed: 0"],inplace=True)
df.head(10)
```

```python
#Round Off the values to 2 decimal places
df['Screen_Size_cm']=np.round(df['Screen_Size_cm'],2)
df.head()
```

```python
#Find the most common Screen Size value
# comm=df['Screen_Size_cm'].value_counts()
comm=df['Screen_Size_cm'].value_counts().idxmax()
comm

```

```python
#replace the missing values with most common value
df['Screen_Size_cm']=df['Screen_Size_cm'].replace(np.nan,comm)
df.tail(10)
```

```python
df.info()  #Still missing values in Weight_kg
#Find the mean of Weight_kg
mean_wt=df['Weight_kg'].astype(float).mean(axis=0)
print("Average Weight in kg:", mean_wt)
```

```python
#Replace the missing values with most common value
df['Weight_kg']=df['Weight_kg'].replace(np.nan,mean_wt)
df.info()  #No missing values now
```

```python
df.tail(10)
```

```python
#Correct Data Format(Transforming Data Types)
df.dtypes #All features have correct data types
```

1 inch = 2.54 cm
1 kg   = 2.205 pounds

```python
#Standardizing the column names
df.head()
#Since usually screen size would be in inches,conersion into inches
df['Screen_Size_cm']=df['Screen_Size_cm']/2.54
df.rename(columns={"Screen_Size_cm":"Screen_Size_inch"},inplace=True)
#convert the weight in Pounds
df['Weight_kg']=df['Weight_kg']*2.205
df.rename(columns={"Weight_kg":"Weight_Pounds"},inplace=True)
df
```

### Data Normalization
Often it is required to normalize a continuous data attribute.Normalize the "CPU_frequency" attribute with respect to the maximum value available in the dataset.

```python
df['CPU_frequency']=df['CPU_frequency']/df['CPU_frequency'].max()
df.head()
```

### Binning
Binning is a process of creating a categorical attribute which splits the values of a continuous data into a specified number of groups. In this case, write a code to create 3 bins for the attribute "Price". These bins would be named "Low", "Medium" and "High". The new attribute will be named "Price-binned".

```python
bins=np.linspace(min(df['Price']),max(df['Price']),4)
groups=["Low","Medium","Max"]
df["Price_Binned"]=pd.cut(df['Price'],bins,labels=groups,include_lowest=True)
df.tail(10)
```

```python
#Visualize these Bins
import matplotlib.pyplot as plt

# For bar pot
plt.bar(groups, df['Price_Binned'].value_counts())
plt.xlabel('Price Binned')
plt.ylabel('Count')
plt.title('Price Binned Bar Plot')
```

### Indicator variables
Converting the "Screen" attribute of the dataset into 2 indicator variables, "Screen-IPS_panel" and "Screen-Full_HD". And Then droping the "Screen" attribute from the dataset.

```python
#Indicators
# df['Screen'].head()
dummies=pd.get_dummies(df['Screen'])
dummies.head()

```

```python
dummies.rename(columns={"Full HD":"Screen_Full_HD","IPS Panel":"Screen_IPS_Panel"},inplace=True)
df=pd.concat([df, dummies], axis=1)
df.drop('Screen', inplace=True, axis=1)
df.head()
```

```python
df.to_csv("Wraggled_Laptops_Data.csv")
```
