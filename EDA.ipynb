### Exploratory Data Analysis of Laptop Pricing Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
file=r"Lab\laptops.csv"
df=pd.read_csv(file)
df.head()
```

#### Step1 - Visualization of Each Feature Patterns


- Continues Features
    1. Screen_Size_inch
    2. CPU_frequency
    3. Weight_pounds

```python
#Regression Plot of these three indidual features with the target variable Price to check the corelation between them.
#1. Screen_Size_inch
sns.regplot(x=df["Screen_Size_inch"], y=df["Price"])
plt.ylim(0,)
```

```python

sns.regplot(x=df["CPU_frequency"], y=df["Price"])
plt.ylim(0,)
```

```python
# 3.Weight_pounds
sns.regplot(x=df["Weight_Pounds"], y=df["Price"])   
plt.ylim(0,)
```

```python
#Corelation Matrix
df[["Screen_Size_inch", "Price"]].corr()
for params in ["Screen_Size_inch", "CPU_frequency", "Weight_Pounds"]:
    print(f"Correlation between {params} and Price: {df[[params, 'Price']].corr()}")  
```

Interpretation: "CPU_frequency" has a 36% positive correlation with the price of the laptops. The other two parameters have weak correlation with price.
